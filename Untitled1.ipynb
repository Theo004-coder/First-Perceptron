```python
import numpy as np

def sigmoid(x):
  return 1/(1- np.exp(-x))


def sigmoid_derivative(x):
  x = sigmoid(x)
  return x*(1-x)

def relu(x):
  return np.maximum(0,x)

def relu_derivative(x):
  return (x>0).astype(float)

def binary_crossentropy(y_true, y_pred):
  eps = 1e-15
  y_pred = np.clip(y_pred, eps, 1-eps)
  return -np.mean(y_true*np.log(y_pred)+ (1 - y_pred) * np.log(1 - y_pred))

def bynary_crossentropy_derivative(y_true, y_pred):
  eps = 1e-15
  y_pred= np.clip(y_pred, eps, 1- eps)
  return (y_pred - y_true)/ (y_pred * (1- y_pred)) / y_true.shape[0]
```

```python
train = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

lable = np.array([
    [1],
    [0],
    [0],
    [1]
])
```

```python
np.random.seed(42)

input_size = 2
hidden_size = 4
output_size = 1

w_1 = np.random.rand(input_size,hidden_size)
b_1 =  np.random.rand(1,hidden_size)
w_2 =  np.random.rand(hidden_size,output_size)
b_2 = np.random.rand(1,output_size)

print(w_1,w_2,b_1,b_2)
```

```python
learning_rate = 0.1

epoch = 1000
for e in range(epoch):
  Z1 = train @ w_1 + b_1
  A1 = relu(Z1)
  Z2 = A1 @ w_2 + b_2
  A2 = sigmoid(Z2)

  loss = binary_crossentropy(lable, A2)

  dA2 = bynary_crossentropy_derivative(lable, A2)
  dZ2 = sigmoid_derivative(Z2) * dA2
  dw2 = A1.T @ dZ2
  db2 = np.sum(dZ2, axis=0, keepdims=True)

  dA1 = dZ2 @ w_2.T
  dZ1 = dA1 * relu_derivative(Z1)
  dw1 = train.T @ dZ1
  db1 = np.sum(dZ1, axis=0, keepdims=True)


  w_1 -= learning_rate*dw1
  b_1 -= learning_rate *db1
  w_2 -= learning_rate * dw2
  b_2 -= learning_rate * db2


  print(f"w1:{w_1}, b1:{b_1}")
  print(f"w2:{w_2}, b2:{b_2}")
  print(loss)
```

```python
def predict(new_x):
    z1 = np.dot(train, w_1)+ b_1
    A1 = relu(z1)
    z2 = np.dot(A1, w_2) + b_2
    A2 = sigmoid(z2)
    return (A2 > 0.5).astype(int)

predict(train)
```
